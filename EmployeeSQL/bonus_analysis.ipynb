```python
from sqlalchemy import create_engine
from config import user
from config import password
engine = create_engine(f'postgresql://{user}:{password}@localhost:5432/SQL-Challenge')
conn = engine.connect()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_sql("SELECT salary from salaries", conn)
```

```python
salaries = data['salary']
```

```python
# Create Histogram of Salaries
plt.hist(salaries)
plt.xlabel('Salary Range')
plt.ylabel('Number of Employees')
plt.show()
```

```python
data1 = pd.read_sql("SELECT * from viewtest3", conn)
```

```python
sort1 = data1.sort_values('Average Salary', ascending=False)
```

```python
# Create Bar Chart of Average Salaries by Title
legit = sort1.set_index('title')

plot1 = legit.plot(kind='bar')
plt.ylabel('Average Salary')
plot1.set_xticklabels(legit.index, rotation=60)

plt.show()
```
